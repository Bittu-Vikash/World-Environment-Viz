```python
print(aa.isnull().sum())
sb.heatmap(aa.isnull(), cbar=False)
plt.show()
```

```python
wt = aa.melt(
    id_vars=['Area', 'Element', 'Months', 'Unit'],
    value_vars=[col for col in aa.columns if col.startswith('Y')],
    var_name='Year',
    value_name='Value'
)

wt['Year'] = wt['Year'].str.replace('Y', '').astype(int)
wt
```

```python
print(wt.isnull().sum())
sb.heatmap(wt.isnull(), cbar=False)
plt.show()
```

```python
wt['Value'].mean()
```

```python
wt['Value'] = wt['Value'].fillna(wt['Value'].median())
```

```python
sb.heatmap(wt.isnull(), cbar=False)
plt.show()
```
